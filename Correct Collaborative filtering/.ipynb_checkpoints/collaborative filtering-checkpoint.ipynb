```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('rating_final.csv', sep=',',skiprows=1, names=['userID','placeID','rating'])
```

```python
df.head()
```

```python
place_titles = pd.read_csv('geoplaces2.csv')
place_titles.head()
```

```python
df = pd.merge(df, place_titles, on='placeID')
df.head()
```

```python
df.describe()
```

```python
ratings = pd.DataFrame(df.groupby('placeID')['rating'].mean())
ratings.head()
```

```python
ratings['number_of_ratings'] = df.groupby('placeID')['rating'].count()
ratings.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
ratings['rating'].hist(bins=50)
```

```python
ratings['number_of_ratings'].hist(bins=60)
```

```python
import seaborn as sns
sns.jointplot(x='rating', y='number_of_ratings', data=ratings)
```

```python
place_matrix = df.pivot_table(index='userID', columns='name', values='rating')
place_matrix.head()
```

```python
ratings.sort_values('number_of_ratings', ascending=False).head(10)
```

```python

#place_matrix
AFO_user_rating =place_matrix['Cafeteria y Restaurant El Pacifico']
contact_user_rating = place_matrix['Restaurant El Muladar de Calzada']
```

```python
AFO_user_rating.head()
contact_user_rating.head()
```

```python
similar_to_air_force_one=place_matrix.corrwith(AFO_user_rating)
```

```python
similar_to_air_force_one.head()
```

```python
similar_to_contact = place_matrix.corrwith(contact_user_rating)
```

```python
similar_to_contact.head()
```

```python
corr_contact = pd.DataFrame(similar_to_contact, columns=['Correlation'])
corr_contact.dropna(inplace=True)
corr_contact.head()
corr_AFO = pd.DataFrame(similar_to_air_force_one, columns=['correlation'])
corr_AFO.dropna(inplace=True)
corr_AFO.head()
```

```python
corr_AFO = corr_AFO.join(ratings['number_of_ratings'])
corr_contact = corr_contact.join(ratings['number_of_ratings'])
corr_AFO .head()
corr_contact.head()
```

```python
corr_AFO[corr_AFO['number_of_ratings'] > 100].sort_values(by='correlation', ascending=False).head(10)
```

```python
corr_contact[corr_contact['number_of_ratings'] > 100].sort_values(by='Correlation', ascending=False).head(10)
```
